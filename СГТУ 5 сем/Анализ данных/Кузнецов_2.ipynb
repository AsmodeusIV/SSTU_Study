apply()

```python
import pandas as pd
# create a series
s = pd.Series([1, 2, 3])
# apply a function to each element in the series
s = s.apply(lambda x: x * 2)
print(s)
```

```python
import pandas as pd

# Создаем DataFrame с названиями фруктов
df = pd.DataFrame({'fruits': ['apple', 'banana', 'cherry', 'date']})

# Применяем функцию apply для преобразования в верхний регистр
df['fruits_upper'] = df['fruits'].apply(lambda x: x.upper())

print(df)
```

transform

```python
import pandas as pd
s = pd.Series([1, 2, 3])
def add_one(x):
    return x + 1
s.apply(add_one)
```

```python
import pandas as pd

# Создаем DataFrame с данными о продажах
df = pd.DataFrame({'store': ['A', 'A', 'B', 'B'],
                   'sales': [100, 200, 150, 250]})

# Нормализуем продажи по каждому магазину
df['normalized_sales'] = df.groupby('store')['sales'].transform(lambda x: (x - x.mean()) / x.std())

print(df)
```

drop()

```python
import pandas as pd
# создадим фрейм данных
df = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
# отбросим колонку 'B'
df = df.drop(columns=['B'])
print(df)
```

```python
import pandas as pd

# Создаем DataFrame с данными
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]})

# Удаляем столбец 'B'
df_dropped = df.drop(columns=['B'])

print(df_dropped)
```

drop_duplicates()

```python
import pandas as pd
# создадим серию с дублирующимися значениями
s = pd.Series([1, 2, 2])
# удалим дубликаты
s = s.drop_duplicates()
print(s)
```

```python
import pandas as pd

# Создаем DataFrame с дубликатами
df = pd.DataFrame({'A': [1, 1, 2, 3],
                   'B': [4, 4, 5, 6]})

# Удаляем дубликаты по столбцу 'A'
df_unique = df.drop_duplicates(subset=['A'])

print(df_unique)
```

```python
import pandas as pd

N = int(input("Введите количество чисел: "))
numbers = [float(input()) for _ in range(N)]

series = pd.Series(numbers)

mean_value = series.mean()

result = series / mean_value

print("Ответ:")
print(result)
```
