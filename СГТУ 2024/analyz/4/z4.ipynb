## Преобразование данных



```python
import pandas as pd # импортировали библиотеку pandas и назвали ее pd
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.columns # список столбцов
```

переименуем названия столбцов, чтобы убрать пробелы

```python
df = df.rename(columns=
                   {'parental level of education':'parental_level_of_education',
                    'test preparation course':'test_preparation_course',
                    'math score':'math_score',
                    'reading score':'reading_score',
                    'writing score':'writing_score'})
df.head()
```

Иногда нужно создать новый признак из уже существующих, например, нам интересно, сколько всего баллов набрал каждый учащийся - просто сложим столбцы с оценкамии поместим сумму в новый столбец all_score. Такая процедура называет broadcasting.

```python
df['all_score'] = df['math_score'] + df['reading_score']+df['writing_score']
df.head()
```

```python
df.info()
```

Видим, что в столбце test_preparation_course есть пропущенные значения.

Полезно посмотреть, какие значения содержатся в столбце. Для категориальных данных можно посчитать частоту встречаемости каждого значения с помошью функции value_counts:

```python
df['gender'].value_counts()
```

```python
df['race/ethnicity'].value_counts()
```

```python
df['parental_level_of_education'].value_counts()
```

Заменить **some high school**  на **high school**

Это можно сделать с помощью анонимной функции и метода  apple


```python
df['parental_level_of_education'] = df['parental_level_of_education'].apply(lambda x: 'high school' if x == 'some high school' else x)
df['parental_level_of_education'].value_counts()
```

```python
df['test_preparation_course'].value_counts() # Пропущенные значения заменить на none
```

```python
print(df['writing_score'].min())
print(df['writing_score'].max())
print(df['writing_score'].mean())
```
