Задание 14
14.	Отобрать количество пассажиров первого класса, которые сели на борт в Саутгемптоне, используя метод query

```python
df.rename(columns={"Passenger class":"Pclass"}, inplace=True)

count_first_class_southampton = df.query("Pclass == 1 and Embarked == 'S'").shape[0]
(count_first_class_southampton)
```

```python
mean_values = df.mean(numeric_only=True)
variance_values = df.var(numeric_only=True)
print("\nСредние значения для количественных переменных:")
print(mean_values)
print("\nДисперсия для количественных переменных:")
print(variance_values)
df.describe()
```
